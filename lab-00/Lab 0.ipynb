## Python functions

A function definition permits the assignement of a name to a statement or a group of statements, just as variable definition permits the assigning of a name to a value.

user-defined functions provide a higher-level abstraction / semantic unit. This is critical for readable, maintanable programs.

````python
def say_stuff():        # define function
  print("some day")     # spaces at the start (typically 2 or 4)
  print("when the world is cold")
  print("I will feel a glow")
  print("just like thinking of you")
  print("and the way you look tonight") # to end function add newline (press return)

# new code that is outside of function
````

````python
say_stuff()
````

## Boolean values and control flow

A boolean value is either True or false.
Booleans are used principally for `if` and `while` statements.

```python
x = False # Boolean literal
it_is_raining = (current_precipitation() == 'rain')
if (it_is_raining): # using a boolean in a conditional
  deploy_umbrella()
else:
  don_t_shirt()

while(current_precipitation() == 'rain'):
  stare_out_window()
go_outside()
```

All values can be converted to / evaluated as Boolean and used in conditionals. All values are True expect None, any number evaluating to 0, and any empty collection (including an empty string)

## Loops and Lists

Transforming a list:

````python
my_list = [5,6,7,8,9,10]
````

````python
my_list = list(range(5, 11)) # (list contents identical
````

````python
new_list = []
for i in range(len(my_list)):        # traditional
  new_list.append(my_list[i] * 2)
````

````python
new_list = [2*x for x in my_list]    # More Pythonic
````

Summing (reducing) a list:

````python
my_sum = 0
for i in range(len(my_list)):        # traditional
  my_sum = = my_sum + my_list[i]
````

````python
my_sum = 0
for e in my_list:                    # More Pythonic
  my_sum = my_sum + e
````

````python
my_sum = sum(my_list)                # (as it happens there is a built-in 'sum' function)
````
